# Data Cleaning
Column split  
Drop null

```python
train[['group','serial']] = train.PassengerId.str.split("_",expand=True)
train
```

```python
test[['group','serial']] = test.PassengerId.str.split("_",expand=True)
test
```

```python
train[['deck','num','side']] = train.Cabin.str.split("/",expand=True)
train
```

```python
test[['deck','num','side']] = test.Cabin.str.split("/",expand=True)
test
```

```python
train = train.dropna(axis=0)
test = test.dropna(axis=0)
```

```python
# get the dummies and store it in a variable
dummies1 = pd.get_dummies(train.side)
dummies2 = pd.get_dummies(train.deck)
dummies3 = pd.get_dummies(train.Destination)
dummies4 = pd.get_dummies(train.HomePlanet)
# Concatenate the dummies to original dataframe
merged = pd.concat([train, dummies1], axis='columns')
merged = pd.concat([merged, dummies2], axis='columns')
merged = pd.concat([merged, dummies3], axis='columns')
merged = pd.concat([merged, dummies4], axis='columns')
 
# drop the values
train_2 = merged.drop(['deck','T','Destination','55 Cancri e','HomePlanet','Mars'], axis='columns')

train_2.columns
```

```python
# get the dummies and store it in a variable
dummies1 = pd.get_dummies(test.side)
dummies2 = pd.get_dummies(test.deck)
dummies3 = pd.get_dummies(test.Destination)
dummies4 = pd.get_dummies(test.HomePlanet)
# Concatenate the dummies to original dataframe
merged = pd.concat([test, dummies1], axis='columns')
merged = pd.concat([merged, dummies2], axis='columns')
merged = pd.concat([merged, dummies3], axis='columns')
merged = pd.concat([merged, dummies4], axis='columns')
 
# drop the values
test_2 = merged.drop(['deck','T','Destination','55 Cancri e','HomePlanet','Mars'], axis='columns')

test_2.columns
```

```python
train.HomePlanet.unique()
```

## Parameter selection

```python
y = train_2.Transported
```

```python
train_2
```

```python
features = ['CryoSleep', 'Age', 'VIP', 'RoomService',
       'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'group', 'serial',
       'num', 'P', 'S', 'A', 'B', 'C', 'D', 'E', 'F', 'G',
       'PSO J318.5-22', 'TRAPPIST-1e', 'Earth', 'Europa']
features
```

```python
x = train_2[features]
```

### Training the Model

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
titanic_model = DecisionTreeRegressor(random_state=1)

# Fit model
titanic_model.fit(x, y)
```

```python
print("Making predictions for the following 5 houses:")
print(x.head())
print("The predictions are")
print(titanic_model.predict(x.head()))
```

```python
from sklearn.metrics import mean_absolute_error

prediction = titanic_model.predict(x)
mean_absolute_error(y, prediction)
```
